```python
import os
from dotenv import load_dotenv, find_dotenv

# find_dotenv() automatically searches parent folders for the .env file
load_dotenv(find_dotenv(), override=True)

# Ensure tracing is explicitly enabled in lower-case
os.environ["LANGCHAIN_TRACING_V2"] = "true"

print("Tracing V2:", os.environ.get("LANGCHAIN_TRACING_V2"))
print("Project:", os.environ.get("LANGCHAIN_PROJECT"))
print("API Key loaded:", bool(os.environ.get("LANGCHAIN_API_KEY")))
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash"
)
print(llm)
```

```python
result = llm.invoke("What is Generative AI?")
print(result)
```

```python
##Prompt Template
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert AI Engineer. Provide me answers based on the questions"),
        ("user","{input}")
    ]
)
```

```python
prompt
```

```python
chain = prompt|llm
response = chain.invoke({"input":"Can you tell me about what is langsmith?"})
```

```python
print(response)
```

```python
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt|llm|output_parser
response = chain.invoke({"input":"What is the purpose of using Langsmith while building generative AI Applications?"})
print(response)
```
