```python
##Web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
webloader = WebBaseLoader(web_path=('https://lilianweng.github.io/posts/2023-06-23-agent/'),bs_kwargs=dict(parse_only = bs4.SoupStrainer(
    class_=("post-title", "post-content", "post-header")
)))
```

```python
webdoc = webloader.load()
print(webdoc)
```
