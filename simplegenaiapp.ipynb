```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
```

```python
loader = WebBaseLoader('https://deepwiki.com/langchain-ai/langsmith-docs/2.1-getting-started')
text = loader.load()
text
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 700, chunk_overlap = 100)
docs = text_splitter.split_documents(text)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embeddings = GoogleGenerativeAIEmbeddings(model = "gemini-embedding-2")
```

```python
vectordb = Chroma.from_documents(docs, embeddings)
```

```python
retriever = vectordb.as_retriever()
```

```python
query = "How to set up an API key on the Langsmith platform?"
```

```python
result = retriever.invoke(query)
print(result[0])
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")
```

```python
##Retrieval chain and document chain 
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_classic.chains import create_retrieval_chain
from langchain_core.prompts import PromptTemplate
```

```python
prompt = PromptTemplate.from_template(
    """
Answer the following questions based only on the provided context:
<context>
{context}
</context>

<question>
{input}
</question>
    """
)
doc_chain = create_stuff_documents_chain(llm, prompt)
```

```python
retrieval_chain = create_retrieval_chain(retriever, doc_chain)
```

```python
retrieval_chain
```

```python
response = retrieval_chain.invoke({"input" : "How to set up an API key on the langsmith platform?"})
```
