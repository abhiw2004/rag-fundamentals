```python
import os
##loading environment variables
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model = "text-embedding-3-large")
embeddings
text = "We are learning OpenAI Embeddings"
query_result = embeddings.embed_query(text)
```

```python
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
```

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
final_docs = splitter.split_documents(docs)
final_docs
```

```python
from langchain_community.vectorstores import Chroma
db = Chroma(final_docs, embeddings_1024)
```

```python
##Retrieve results from query vectorstore db
query = "for the time being, to believe that this is spoken from our hearts"
retreival_result = db.similarity_search(query)
print(retreival_result)
```
