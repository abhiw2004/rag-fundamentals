```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
loader = TextLoader('speech.txt')
docs = loader.load()
```

```python
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap = 30)
docs = text_splitter.split_documents(docs)
```

```python
embeddings = HuggingFaceEmbeddings(model_name = 'all-MiniLM-L6-v2')
db = FAISS.from_documents(docs, embeddings)
```

```python
query = "What must be the peace of the world planted upon?"
search_result = db.similarity_search(query)
```

```python
print(search_result[0])
```

```python
##Using Retriever
retreiver = db.as_retriever()
docs = retreiver.invoke(query)
print(docs[0].page_content)
```

```python
##Similarity search with score
docs_and_score = db.similarity_search_with_score(query)
docs_and_score
```

```python
##Searching by the vector
vector = embeddings.embed_query(query)
vector
```

```python
search_and_score = db.similarity_search_with_score_by_vector(vector)
search_and_score
```

```python
##Saving the vectorstore
db.save_local("faiss_local")
```

```python
new_db = FAISS.load_local('faiss_local',embeddings,allow_dangerous_deserialization=True)
```

```python
result = new_db.similarity_search(query)
```
